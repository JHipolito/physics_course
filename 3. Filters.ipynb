```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# calculate module and phase against impulse response
w,h  = signal.freqz([1/2,1/2],1)
plt.figure(1)
plt.plot(w,20*np.log10(abs(h)))
plt.xlabel('frequency (rad/s)',fontsize=18)
plt.ylabel('Decibels',fontsize=18)
plt.title('Module')
plt.grid()
plt.show()
plt.figure(2)
plt.plot(w,np.angle(h)/np.pi*180)
plt.xlabel('frequency (rad/s)',fontsize=18)
plt.ylabel('Phase (degrees)',fontsize=18)
plt.show()
```
